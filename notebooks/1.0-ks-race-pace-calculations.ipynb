```python
import fastf1 as f1 # f1 data library
import pandas as pd
import numpy as np
```

```python
event_schedule = f1.get_event_schedule(2024)

event_schedule[event_schedule.Session5DateUtc < pd.Timestamp.today()]
```

```python
session = f1.get_session(2024, 1, 'FP1')

session.load()

session1 = f1.get_session(2024, 1, 'R')

session1.load()

session
```

```python
laps = session.laps.dropna(subset=['LapTime'])

ver_laps = laps[laps['Driver'] == 'VER']

ver_laps
```

```python
for driver in laps.Driver.unique():
    print(driver)
    driver_laps = laps[laps['Driver'] == driver]
    for stint in driver_laps.Stint.unique():
        driver_stint_laps = driver_laps[driver_laps['Stint'] == stint]
        print(stint)
        print(len(driver_stint_laps))
        print(np.mean(driver_stint_laps['LapTime']))
    break
```

```python
racelaps = session1.laps.dropna(subset=['LapTime'])

racelaps
```

```python
for driver in racelaps.Driver.unique():
    print(driver)
    driver_laps = racelaps[racelaps['Driver'] == driver]
    for stint in driver_laps.Stint.unique():
        driver_stint_laps = driver_laps[driver_laps['Stint'] == stint]
        print(stint)
        print(len(driver_stint_laps))
        print(np.mean(driver_stint_laps['LapTime']))
    break
```

```python
# This looks good hehe

quick_laps = session.laps.pick_quicklaps().pick_driver("VER")
race_quick_laps = racelaps.pick_quicklaps().pick_driver("VER")

print(f"Practice: {np.mean(quick_laps['LapTime'])}")
print(f"RACE: {np.mean(race_quick_laps['LapTime'])}")


race_quick_laps
```
